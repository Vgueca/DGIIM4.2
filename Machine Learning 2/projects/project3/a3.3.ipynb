```python
CollegeTrain = pd.read_csv("CollegeTrain.csv", index_col=0)
CollegeTest = pd.read_csv("CollegeTest.csv", index_col=0)
```

```python
# Adjusting dataset and the samples
x_train= CollegeTrain.sample(frac = 0.05, random_state=0)

y_train = x_train['Outcome']
x_train = x_train.drop(['Outcome'], axis = 1)

x_test = CollegeTest.sample(n=100, random_state=0)

y_test = x_test['Outcome']
x_test = x_test.drop(['Outcome'], axis = 1)
```

```python
# training the decision tree classifier

clf = DecisionTreeClassifier(random_state=0)
clf = clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred))
```

```python
indiv_test_acc =[]
lower_bound = []
upper_bound = []

mean_test_acc = []
lower_bounds = []
upper_bounds = []
alpha = 2.5
for i in range(20):
    # preprocessing
    x_train= CollegeTrain.sample(frac = 0.05, random_state=i)

    y_train = x_train['Outcome']
    x_train = x_train.drop(['Outcome'], axis = 1)

    x_test = CollegeTest.sample(n=100, random_state=i)

    y_test = x_test['Outcome']
    x_test = x_test.drop(['Outcome'], axis = 1)

    # training the decision tree classifier
    clf = DecisionTreeClassifier(random_state=0)
    clf = clf.fit(x_train, y_train)
    y_pred = clf.predict(x_test)
    
    #individual preccision
    test_acc = accuracy_score(y_test, y_pred)
    indiv_test_acc.append(test_acc)
    
    #bounds of confidence interval for the test_acc
    lower_bound_aux = test_acc - 1.96*(((1-test_acc)*test_acc)/100)**(1/2)
    upper_bound_aux = test_acc + 1.96*(((1-test_acc)*test_acc)/100)**(1/2)
    lower_bound.append(lower_bound_aux)
    upper_bound.append(upper_bound_aux)
    
    #observed bounds and mean_test_Acc 
    test_accs = []
    for j in range(100):
        x_test = CollegeTest.sample(n=100, random_state=j)
        y_test = x_test['Outcome']
        x_test = x_test.drop(['Outcome'], axis = 1)
        y_pred = clf.predict(x_test)
        test_accs.append(accuracy_score(y_test, y_pred))
    
    mean_test_acc.append(sum(test_accs)/len(test_accs))
    lower_bounds.append(np.percentile(test_accs, alpha))
    upper_bounds.append(np.percentile(test_accs, 100-alpha))
```

```python
frame = pd.DataFrame(columns = ["indiv_test_acc", "CI_lower_bound", "CI_upper_bound", "mean_test_acc", "observed_lower_bound", "observed_upper_bound"])
frame["indiv_test_acc"] = indiv_test_acc
frame["CI_lower_bound"] = lower_bound
frame["CI_upper_bound"] = upper_bound
frame["mean_test_acc"] = mean_test_acc
frame["observed_lower_bound"] = lower_bounds
frame["observed_upper_bound"] = upper_bounds
```
